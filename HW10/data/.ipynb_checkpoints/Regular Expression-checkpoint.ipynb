```python
contents = open("yellowpages.txt").read()
# Show me characters between 0 and 500
contents[0:500]
```

```python
# re is regular expressions
import re
```

```python
re.findall(r"(.*?)[\s.]*(\d\d\d-\d\d\d-\d\d\d\d)", contents)
```

```python
pd.DataFrame(re.findall(r"(.*?)[\s.]*(\d\d\d-\d\d\d-\d\d\d\d)", contents))
```

```python
pride = open("Austen_Pride.txt").read()
pride[0:500]
```

```python
len(re.findall(r"\b[Ss]he\b", pride))
```

```python
len(re.findall(r"\b[Hh]e\b", pride))
```

```python
f_verbs = re.findall(r"\b[Ss]he (\w*)", pride)
f_verbs
```

```python
from collections import Counter
```

```python
Counter(f_verbs).most_common(100)
```
