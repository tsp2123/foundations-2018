```python
#Unionization Rates from 2000 - 2016 Total

regular_data = driver.find_elements_by_class_name('regular-data')
```

```python

regular_data[2].text.split('\n')[7]
```

```python
regular_data[0].find_elements_by_tag_name('pre')[0].text
```

```python
Union_Information = []

for elements in regular_data[::2]:
    row = {}
    keys = elements.text.split('\n')
    print(keys)
#     print('------')
    series_id = keys[1]
    series_title = keys[2]
    union = keys[3]
    industry = keys[4]
    occupations = keys[5]
    sex = keys[6]
    race = keys[7]
    ethnic_origin = keys[8]
    age = keys[9]
    earnings = keys[10]
    class_of_workers = keys[11]
    labor_force_status = keys[12]
    
    row['series_id'] = series_id
    row['series_title'] = series_title
    row['union'] = union
    row['industry'] = industry
    row['occupations'] = occupations
    row['sex'] = sex
    row['race'] = race
    row['ethnic_origin'] = ethnic_origin
    row['age'] = age
    row['earnings'] = earnings
    row['class_of_workers'] = class_of_workers
    row['labor_force_status'] = labor_force_status
    
    Union_Information.append(row)
    
Union_Information
    

    
    
```

```python
import pandas as pd

pd.DataFrame(Union_Information)
```
