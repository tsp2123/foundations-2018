<h1> Can I put this all into one dictionary?<h1>

###Here's what I'm doing here: I'm gonna make a dictionary with a Key for my States. Then within each state will be another key for my years and in that we shall have our minimum wage data###

```python


Ultimate_Loop = driver.find_elements_by_tag_name('table')[2].find_elements_by_tag_name('tr')
Ultimate_Loop2 = driver.find_elements_by_tag_name('table')[3].find_elements_by_tag_name('tr')
Ultimate_Loop3 = driver.find_elements_by_tag_name('table')[4].find_elements_by_tag_name('tr')

#The easiest way to get this data presentable is by making a dictionary in a dictionary. So I'm skipping making the list
row = {}


#This loop is for data from 2000 to 2006

for iterables in Ultimate_Loop[1:]:

    
    states = iterables.find_elements_by_tag_name('td')[0].text
    row[states] = {}
#Over here I'm specifying that the dictionary row has a key 'states' which is currently an empty dictionary 
#The following steps will populate this dictionary within a dictionary and for whatever reason I dont have to keep making this dictionary in my other loops
    
    Wage2000 = iterables.find_elements_by_tag_name('td')[1].text
    row[states]['2000'] = Wage2000
    
    Wage2001 = iterables.find_elements_by_tag_name('td')[2].text
    row[states]['2001'] = Wage2001
    
    Wage2002 = iterables.find_elements_by_tag_name('td')[3].text
    row[states]['2002'] = Wage2002
    
    Wage2003 = iterables.find_elements_by_tag_name('td')[4].text
    row[states]['2003'] = Wage2003
    
    Wage2004 = iterables.find_elements_by_tag_name('td')[5].text
    row[states]['2004'] = Wage2004
    
    Wage2005 = iterables.find_elements_by_tag_name('td')[6].text
    row[states]['2005'] = Wage2005
    
    Wage2006 = iterables.find_elements_by_tag_name('td')[7].text
    row[states]['2006'] = Wage2006
    
    
    
#This loop is for data from 2007 to 2013    
    
for iterables in Ultimate_Loop2[1:]:
    
    states = iterables.find_elements_by_tag_name('td')[0].text

    
    Wage2007 = iterables.find_elements_by_tag_name('td')[1].text
    row[states]['2007'] = Wage2007
    
    Wage2008 = iterables.find_elements_by_tag_name('td')[2].text
    row[states]['2008'] = Wage2008
    
    Wage2009 = iterables.find_elements_by_tag_name('td')[3].text
    row[states]['2009'] = Wage2009
    
    Wage2010 = iterables.find_elements_by_tag_name('td')[4].text
    row[states]['2010'] = Wage2010
    
    Wage2011 = iterables.find_elements_by_tag_name('td')[5].text
    row[states]['2011'] = Wage2011
    
    Wage2012 = iterables.find_elements_by_tag_name('td')[6].text
    row[states]['2012'] = Wage2012
    
    Wage2013 = iterables.find_elements_by_tag_name('td')[7].text
    row[states]['2013'] = Wage2013
    

    
    

#This loop is for data from 2014 to 2016
    
    
for iterables in Ultimate_Loop3[1:-2]:
    
    states = iterables.find_elements_by_tag_name('td')[0].text
    print(states)
    
    Wage2014 = iterables.find_elements_by_tag_name('td')[1].text
    row[states]['2014'] = Wage2014
    print(row[states]['2014'])
    
    Wage2015 = iterables.find_elements_by_tag_name('td')[2].text
    row[states]['2015'] = Wage2015
    
    Wage2016 = iterables.find_elements_by_tag_name('td')[3].text
    row[states]['2016'] = Wage2016
    

    
    
    
# Minimum_Wage_Data    


```
